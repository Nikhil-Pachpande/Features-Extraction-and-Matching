```python
# Import the necessary libraries and packages
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
```

```python
# read the image
image1 = cv2.imread(os.path.join('ud.jpg'))
```

```python
# Convert the image to RGB
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
```

```python
# Initialize the SIFT Detector
orb = cv2.ORB_create()
```

```python
# Finding the key points and descriptors with SIFT
key_points1, descriptors1 = orb.detectAndCompute(image1, None)
```

```python
# Draw the keypoints on the Image
image_key_points1 = cv2.drawKeypoints(image1, key_points1, image1, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
# Display the ORB interest points on the Image
plt.figure(figsize=(16,16))
plt.title('ORB Interest Points')
plt.imshow(image_key_points1)
plt.show()
```
