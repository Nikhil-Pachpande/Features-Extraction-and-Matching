```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
```

```python
image = cv2.imread(os.path.join('ud.jpg'))
```

```python
cv2.imshow('Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
orb = cv2.ORB_create()
```

```python
key_points, description = orb.detectAndCompute(image, None)
```

```python
image_key_points = cv2.drawKeypoints(image, key_points, image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.figure(figsize=(16,16))
plt.title('ORB Interest Points')
plt.imshow(image_key_points)
plt.show()
```
